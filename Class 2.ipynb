```python
#Fibbonacci series using while loop
a = 0
b = 1
n = 10 # Specifies the output range i.e upto 10
while n > 0:
    c = a + b
    a = b
    b = c
    n = n-1
    print(c)
```

```python
x = (1, 2, 3, 4,5,6)
y = (10, 20, 30)
z = (100, 200)

n = max(len(x), len(y), len(z))

if len(x) < n:
    diff = n - len(x)
    x = x + (1,) * diff
    print("x:", x)
if len(y) < n:
    diff = n - len(y)
    y = y + (1,) * diff
    print("y:", y)
if len(z) < n:
    diff = n - len(z)
    z = z + (1,) * diff
    print("z:", z)
    
for i in range(n):
    print(x[i]*y[i]*z[i])
```

```python
def pascal_triangle(n):
    triangle = []  # will hold all rows

    for row in range(n):
        if row == 0:
            triangle.append([1])  # first row is just [1]
        else:
            prev = triangle[-1]   # previous row
            new_row = [1]         # start with 1
            # fill middle values
            for i in range(1, row):
                new_row.append(prev[i-1] + prev[i])
            new_row.append(1)     # end with 1
            triangle.append(new_row)

    return triangle

# Example: print 6 rows
rows = pascal_triangle(6)
for r in rows:
    print(r)
```
