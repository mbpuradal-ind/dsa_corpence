```python
# Very important example
x = '''hello, we are learning python, Python is very popular, 
at the end of the class I am sure I know that Python,
is not an animal'''
print(x.split())
char_idx = 0
for idx, word in enumerate(x.split()):
    if 'python' in word.lower():
        print(char_idx)
    char_idx += len(word) + 1
```

```python
x = '''hello, we are learning python, Python is very popular, 
at the end of the class I am sure I know that Python,
is not an animal'''
print(x.split())
char_idx = 0
for idx, word in enumerate(x.split()):
    if len(word) >= 3 and (word[0] in 'aeiou' or word[-1] in 'aeiou'):
        print(char_idx) 
    char_idx += len(word) + 1
```

Regular expressions

```python
text = ''' We are learning python, Python is the most popular programming language, are you learning python?'''
#import question in interviews
#check below code by removing \b, also without \w+, this gives an idea on the usage of regular expressions

import re
pattern = re.compile(r'\b[aeiou]\w+[aeiou]\b', flags = re.IGNORECASE)
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

```python
text = ''' 
https://www.google.com
http://tesla.co.ai
www.yahoo.ai
https://www.microsoft.org
http//:www.google.co
this is a website
999-876-9874
abc@gmail.com
abc.india@outlook.co.in
'''
#? = means anything befor that is optional
#pattern = re.compile(r'(https?://(?:www\.)?[\w-]+(?:\.[a-zA-Z]{2,}+', flags = re.IGNORECASE) 
pattern = re.compile(r'(\w+://(\w+.?)\w{3})', flags = re.IGNORECASE) 
matches = pattern.finditer(text)
for match in matches:
    print(match)
```
