Arithmetic Operators

```python
## Example of Arithmetic Operators >> Used for basic mathematical operations.
x=5
y=3
print(x+y) #Adds two operands
print(x-y) #Subtracts the second operand from the first
print(x*y) #Multiplies two operands
print(x/y) #Divides the first operand by the second, returning a float
print(x//y) #Floor Division (//):Divides the first operand by the second, returning an integer
print(x%y) #Modulus (%) : Returns the remainder when the first operand is divided by the second
print(x**y) #Exponentiation (**): Raises the first operand to the power of the second
```

Comparison (Relational) Operators

```python
# Example of Comparison Operators >> Used to compare values; returns True or False.
x = 10
y = 20
print(x == y) # Equal
print(x != y) # Not equal
print(x > y) # Greater than
print(x < y) # Less than
print(x >= y) # Greater or equal
print(x <= y) #	Less or equal
```

Logical Operators

```python
# Example of Logical Operators >> Used to combine conditional statements.
a = True
b = False
print(a and b) # AND: True if both are True
print(a or b) # OR: Returns true if either condition is true
print(not a) # NOT: Inverts the result
print(not b)
```

Bitwise Operators

```python
# Example of Bitwise Operators >> Operate on binary representations of integers.
# Example (a=5 (0101), b=3 (0011))
# AND (&): a & b → 1 (0001)
# OR (`): `a
# XOR (^): a ^ b → 6 (0110)
# NOT (~): ~a → -6
# Left shift (<<): a << 1 → 10 (1010)
# Right shift (>>): a >> 1 → 2 (0010)

a = 5 #(0101)
b = 3 #(0011)
print(a & b)
```

Identity Operators

```python
# Example of Identity Operators >> Compare memory locations (object identity).
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b) # True, because b is the same object as a
print(a is not c) # True, because c is not the same object as a
print(a == c) # == means "is equal"
```

Boolean operator

```python
# Examples of Boolean operators in Python

a = True
b = False

print(a and b)   # and = both must be True
print(a or b)    # or = at least one is True
print(not a)     # not = inverts or reverses value, in present ex a=True so it is reversed to False 

# Using with comparisons
x = 10
y = 5

print(x > 5 and y < 10)  # True  (both conditions are True)
print(x < 5 or y < 10)   # True  (second condition is True)
print(not (x == 10))     # False (x == 10 is True, so not makes it False)
```
